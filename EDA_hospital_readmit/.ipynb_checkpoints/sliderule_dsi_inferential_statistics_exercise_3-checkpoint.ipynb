# Hospital Readmissions Data Analysis and Recommendations for Reduction

### Background
In October 2012, the US government's Center for Medicare and Medicaid Services (CMS) began reducing Medicare payments for Inpatient Prospective Payment System hospitals with excess readmissions. Excess readmissions are measured by a ratio, by dividing a hospital’s number of “predicted” 30-day readmissions for heart attack, heart failure, and pneumonia by the number that would be “expected,” based on an average hospital with similar patients. A ratio greater than 1 indicates excess readmissions.

```python
def pearson_r(x, y):
    """Compute Pearson correlation coefficient between two arrays."""
    # Compute correlation matrix: 
    corr_mat = np.corrcoef(x, y)
    # Return entry [0,1]
    return corr_mat[0,1]
```

```python
df = clean_hospital_read_df
df.drop(columns='Footnote', inplace=True)
df.dropna(how='any', inplace=True)
```

```python
df['Readmission per discharge'] = df['Number of Readmissions'] / df['Number of Discharges']
```

```python
df_low = df[df['Number of Discharges'] <= 100]
df_medium = df[(df['Number of Discharges'] > 100) & (df['Number of Discharges'] <= 1000)]
df_high = df[df['Number of Discharges'] > 1000]
```

```python
def bootstrap_replicate_1d(data, func):
    bs_sample = np.random.choice(data, size=len(data))
    return func(bs_sample)

def draw_bs_reps(data1, func, size=1):
    bs_replicates = np.empty(size)
    for i in range(size):
        bs_replicates[i] = bootstrap_replicate_1d(data, func)
    return bs_replicates

def diff_of_means(data1, data2):
    return (np.mean(data1) - np.mean(data2))
```

```python
x = df_low['Excess Readmission Ratio']
y = df_medium['Excess Readmission Ratio'].append(df_high['Excess Readmission Ratio'])
original_diff_means = diff_of_means(x, y)
```

```python
# Hypothesis: mean excess readmission ratio is same for small and large hospitals
# Bootstrap test 

original_diff_means = diff_of_means(x, y)

# Initialize bootstrap replicates: bs_replicates
bs_replicates = np.empty(10000)

for i in range(10000):
    # Generate bootstrap sample
    bs_sample = np.random.choice(df['Excess Readmission Ratio'], size=len(df))
    
    # Compute replicate
    bs_replicates[i] = diff_of_means(bs_sample[:len(x)],
                                     bs_sample[len(y):])

# Compute and print p-value: p
p = np.sum(bs_replicates >= original_diff_means) / len(bs_replicates)

# Variance for two samples, white and black
var_2samp = (x.var()/len(x)) + (y.var()/len(y))

# Standard error
std_dev = np.sqrt(var_2samp)

# t statistic
t_stat = (x.mean() - y.mean()) / std_dev

# Margin of error with 95% confidence
margin_error = 1.96 * std_dev

# Confidence interval
conf_interval = original_diff_means + np.array([-1, 1]) * margin_error

print('BOOTSTRAP')
print('\nDifference in excess readmission rate means small/large: ', original_diff_means)
print('T statistics: ', t_stat)
print('pvalue: ', p)
print('Margin of error: ', margin_error)
print('Confidence interval: ', conf_interval)
```

```python
# Hypothesis: correlation between number of discharges and excess readmission ratio is mere chance
# Permutation test
x = df['Number of Discharges']
y = df['Excess Readmission Ratio']

# Compute observed correlation: r_obs
r_obs = pearson_r(x, y)

# Initialize permutation replicates: perm_replicates
perm_replicates = np.empty(10000)

# Draw replicates
for i in range(10000):
    # Permute illiteracy measurments: illiteracy_permuted
    x_permuted = np.random.permutation(x)

    # Compute Pearson correlation
    perm_replicates[i] = pearson_r(x_permuted, y)


# Compute p-value: p
p = np.sum(perm_replicates <= r_obs) / len(perm_replicates)
print('p-val =', p)
```

```python
# Hypothesis: correlation between number of discharges and readmission per discharge is by random chance

x = df['Number of Discharges']
y = df['Readmission per discharge']

# Compute observed correlation: r_obs
r_obs = pearson_r(x, y)

# Initialize permutation replicates: perm_replicates
perm_replicates = np.empty(10000)

# Draw replicates
for i in range(10000):
    # Permute illiteracy measurments: illiteracy_permuted
    x_permuted = np.random.permutation(x)

    # Compute Pearson correlation
    perm_replicates[i] = pearson_r(x_permuted, y)


# Compute p-value: p
p = np.sum(perm_replicates <= r_obs) / len(perm_replicates)
print('p-val =', p)
```

```python
x = df[df['Number of Discharges'] <= 300]['Excess Readmission Ratio']
y = df[df['Number of Discharges'] > 300]['Excess Readmission Ratio']

t_ttest = stats.ttest_ind(x, y)[0]
p_ttest = stats.ttest_ind(x, y)[1]

print('T - test')
print('\nReadmission rate means - small: ', x.mean())
print('Readmission rate means - big: ', y.mean())
print('Difference in excess readmission rate means small/large: ', original_diff_means)
print('T statistics: ', t_ttest)
print('pvalue: ', p_ttest)
```

**Preliminary report analysis**

It was correctly identified that rate of readmission is trending down with increasing number of discharges. However, it was not backed by any statistical analysis. There is a downward sloping pearson correlation of 0.097 between the size of hospital (assuming proportional to number of discharges) and excessive readmission.  We permuted the excessive readmission rate with number of discharges and not one instance out of 10,000 runs spit correlation as high. Implies, there is **statistically significant negative correlation** between the size of hospital and the number of readmitted patients.

It is very easy to confuse the above analysis with *causation*. Let's discuss this in later part as final analysis.

Above scatter plot somewhat shows in red and green sections, smaller hospitals have more incidents of readmissions in contrast to bigger hospitals. Again, there was no statistical significance provided in preliminary report. 

More than half of small hospitals have excessive readmissions, 63% to be precise. As expected from the correlation discussion, only 44% of big hospitals process excessive readmissions. We ran T test on the **difference in means of excessive readmission rate** between the two groups of hospitals, 0.016, and can confirm there is a **statistically significant difference** with more than 99.99% significance.

*Implication!* that is all we could conclude from this analysis so far, almost 1 out 3 small hospitals have chances of excessive readmissions. Other than the statistically proven correlation.


**Statistical analysis**

We categorized hospitals into two groups, small hospitals with less than 300 patient discharges and large ones with more than 300 discharges. Difference in mean readmission rate is 0.0163. 

$H_0$: The mean readmission rate is same for both small and big hospitals.

$H_A$: There is a statistically significant difference in mean readmission rate for two groups with 0.01% significance level.

T test produced a very small pvalue, which means we can reject the null hypothesis and confirm the statistically significant difference in mean readmission rate between small and big hospitals.

+ T statistics:  8.447991822364562
+ pvalue:  3.319407529146751e-17
