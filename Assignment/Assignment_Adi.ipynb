### Identify pockets in San Francisco with a high concentration of active businesses. Please provide visualization(s) to support your analysis.

#### Analysis

1. Location data is available in column 'Business Location'. Drop rows where location is null.
2. Fetch coordinates from location using regex.
3. Concentration of active businesses is required for San Francisco only. Drop rows where Neighborhood is null since this is provided for SF location only.
4. List Neighborhoods with count. This is the number for active businesses and locations. Plot the data.
5. Scatter plot the coordinates on x/y plane. Areas with high concentration have overlapping points with denser color. 

As seen from the scatter plot in north east section, fidi and mission are the most populated areas. 

```python
# Assign active business/location flags

df['Active_business'] = np.where(df['Business End Date'].isnull(), 1, 0)
df['Active_location'] = np.where(df['Location End Date'].isnull(), 1, 0)

# Drop rows where Business Location is null

df_location = df.dropna(subset=['Business Location'])

# Create new features, latitude and longitude

df_location['lat'] = df_location['Business Location'].str.extract('\(([^()]*)\)', expand=True)\
                                                                    [0].str.split(',', expand=True)[0]

df_location['lon'] = df_location['Business Location'].str.extract('\(([^()]*)\)', expand=True)\
                                                                    [0].str.split(',', expand=True)[1]

# Drop rows where coordinates are not present within parentheses

df_location.dropna(subset=['lat', 'lon'], inplace=True)

# Convert lat/lon from string to numeric values

df_location['lat'] = df_location['lat'].apply(pd.to_numeric, errors='coerce')
df_location['lon'] = df_location['lon'].apply(pd.to_numeric, errors='coerce')

# Rename Neighborhood column for easy access

df_location.rename(columns={'Neighborhoods - Analysis Boundaries': 'Neighborhood'}, inplace=True)

# Filter data for active businesses/locations only

location = df_location[(df_location.Active_business == 1) & (df_location.Active_location == 1)]

# Drop rows for non SF locations. Assumption all SF locations have neighborhood value pre-assigned

location.dropna(subset=['Neighborhood'], inplace=True)

# Drop bad data where coordinates are out of SF bounds

location.drop(location[(location.lat < 37.7) | (location.lat > 37.9)].index, inplace=True)
location.drop(location[(location.lon < -122.52) | (location.lon > -122.3)].index, inplace=True)
```

```python
# List of neighborhoods with active businesses

print('Neighborhoods with most active businesses (top 10):\n')
print(location.Neighborhood.value_counts()[:10])
```

```python
# Horizontal bar plot for top 10 neighborhoods with active locations in SF

fig, ax = plt.subplots(figsize = (10, 5))
location.Neighborhood.value_counts()[:10].plot(kind='barh', ax=ax)

ax.set_xlabel('Total active locations')
ax.set_title('Active businesses/location in SF Neighborhoods (top 10)')
```

```python
# Plot location data for active businesses on scatter plot
# High concentration areas are depicted with overlapping markers with darker color

fig, ax = plt.subplots(figsize = (15, 15))
location.plot(kind="scatter", x="lon", y="lat", alpha=0.2, ax=ax, marker='.')
plt.show()
```
